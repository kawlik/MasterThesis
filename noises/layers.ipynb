# Libraries

```python
# tensorflow utils
import tensorflow as tf

# matplotlib utils
import matplotlib as mpl
import matplotlib.pyplot as plt

# local assets
import layers

# local config
mpl.rcParams['font.family'] = 'monospace'
```

# Visualize :: Cauchy

```python
zeros = tf.zeros((28, 28))
noise = layers.CauchyNoiseLayer(0.1, (-1, +1))(zeros)

plt.imshow(noise, vmin=-1, vmax=+1)
```

# Visualize :: Normal

```python
zeros = tf.zeros((28, 28))
noise = layers.NormalNoiseLayer(0, 0.1, (-1, +1))(zeros)

plt.imshow(noise, vmin=-1, vmax=+1)
```

# Visualize :: Uniform

```python
zeros = tf.zeros((28, 28))
noise = layers.UniformNoiseLayer(-0.1, +0.1, (-1, +1))(zeros)

plt.imshow(noise, vmin=-1, vmax=+1)
```
